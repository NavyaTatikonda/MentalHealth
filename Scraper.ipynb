```python
import requests #requests allow you to send requets to the sites
from bs4 import BeautifulSoup
```

```python
#Source URL 
url="https://www.healthline.com/health/mental-health#disorders"
```

```python
r=requests.get(url)
```

```python
print(r.content)
```

```python
soup = BeautifulSoup(r.content, 'html5lib') 
print(soup.prettify()) 
```

```python
what_MH=soup.find_all(class_='css-0')
#what_MH_p=soup.find_all('p')
#what_MH_p
what_MH
```

```python
li=soup.find('div',{'class':'css-0'})
children = li.findChildren("a" , recursive=False)
for child in children:
    print(child)
```
